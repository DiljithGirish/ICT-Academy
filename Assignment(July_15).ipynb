## Numpy Problems

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("auto-mpg.csv")
```

```python
#Qn.1
mpg_array = np.array(data.mpg)
mpg_array.sort()
mpg_array
```

```python
mean = np.mean(mpg_array)
median = np.median(mpg_array)
std_deviation = np.std(mpg_array)

```

```python
print(f"Mean is {mean}")
print(f"Median is {median}")
print(f"Std deviation is {std_deviation}")
```

```python
#cars with mpg greated than 25

mpg_morethan_25 = np.sum(mpg_array > 25)
print(mpg_morethan_25)
```

```python
#Qn 2

car_list_array = np.array(data[(data['cylinders']>6)]['car name'])
car_names_list = car_list_array.tolist()
print(len(car_names_list))
print(car_names_list)
```

```python
#Qn 3

weight_array = np.array(data.weight)

percentile25 = np.percentile(weight_array, 25)
percentile50 = np.percentile(weight_array, 50)
percentile75 = np.percentile(weight_array, 75)
print("Percentile 25:",percentile25)
print("Percentile 50:",percentile50)
print("Percentile 75:",percentile75)
```

```python
#Qn 4

acceleration_array = np.array(data.acceleration)
min_value = np.min(acceleration_array)
max_value = np.max(acceleration_array)

normalized_data = (acceleration_array - min_value) / (max_value - min_value)
print(acceleration_array)
print()
print(normalized_data)
```

```python
# Qn 5


data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
mean_hp = data['horsepower'].mean()
data['horsepower'] = data['horsepower'].fillna(mean_hp)
hp_array = data['horsepower'].to_numpy()
new_hp_array = hp_array * 1.10
print(new_hp_array)
```

```python
#Qn 6

displacement = np.array(data[data["origin"]==2]["displacement"])
avg_displacement = np.average(displacement)
print(avg_displacement)
```

```python
# Qn 7
car_features = np.column_stack((data["mpg"], data["horsepower"], data["weight"]))
given_vector = np.array([1, 0.5, -0.2])
dot_product_result = np.dot(car_features, given_vector)
print(dot_product_result)
```

```python
#Qn 8

car_names_sorted = (
    data.sort_values("model year", ascending=False)["car name"].head(5).tolist()
)
car_names_sorted
```

```python
#Qn 9

corr_coef = np.corrcoef(mpg_array, weight_array)[0, 1]
print("Pearson correlation coefficient between mpg and weight:", corr_coef)
```

```python
#Qn 10
unique_cylinders = np.unique(data["cylinders"].values)
mean_mpg_by_cyl = {
    cyl: np.mean(mpg_array[data["cylinders"].values == cyl]) for cyl in unique_cylinders
}
mean_mpg_by_cyl
```

##Pandas Problems

```python
#Qn 1

import pandas as pd

df = pd.read_csv("auto-mpg.csv")
display(df.head(10))
print("Shape (rows, columns):", df.shape)
display(df.describe())
```

```python
#Qn 2

filtered_df = df[(df["model year"] == 75) & (df["weight"] < 3000)][
    ["car name", "weight", "mpg"]
]
display(filtered_df)
```

```python
#Qn 3

missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
median_hp = df["horsepower"].median()
df["horsepower"] = df["horsepower"].fillna(median_hp)
print("Median horsepower:", median_hp)
```

```python
#Qn 4
df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]
display(df[["horsepower", "weight", "power_to_weight_ratio"]].head())
```

```python
# Qn 5

mean_mpg_by_origin = df.groupby("origin")["mpg"].mean()
display(mean_mpg_by_origin)
```

```python
#Qn 6

def performance_score(row):
    return row["mpg"] * row["acceleration"] / row["weight"]


top10_mpg = df.sort_values(by="mpg", ascending=False).head(10)
display(top10_mpg[["car name", "mpg", "acceleration", "weight"]])
```

```python
#Qn 7

df["performance_score"] = df.apply(performance_score, axis=1)
display(df[["car name", "mpg", "acceleration", "weight", "performance_score"]].head())
```

```python
#Qn 8
summary_by_year = (
    df.groupby("model year")[["mpg", "weight", "horsepower"]].mean().reset_index()
)
summary_by_year.rename(
    columns={"mpg": "avg_mpg", "weight": "avg_weight", "horsepower": "avg_horsepower"},
    inplace=True,
)
display(summary_by_year)
```

```python
# Qn 9


high_mpg_cars = df[df["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]
high_mpg_cars.to_csv("high_mpg_cars.csv", index=False)
```

```python
# Qn 10
Q1=np.percentile(df['mpg'],25)
Q2=np.percentile(df['mpg'],50)
Q3=np.percentile(df['mpg'],75)
IQR= Q3-Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["mpg"] < lower_bound) | (df["mpg"] > upper_bound)][
    ["car name", "mpg", "model year"]
]
display(outliers)
```

##Matplotlib Problems

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("auto-mpg.csv")
```

```python
#Qn 1
plt.hist(df["mpg"], bins=20, edgecolor="black")
plt.xlabel("Miles per Gallon (mpg)")
plt.ylabel("Frequency")
plt.title("Distribution of MPG in the Dataset")
plt.show()
```

```python
#Qn 2

df.boxplot(column="mpg", by="cylinders", grid=False)
plt.xlabel("Number of Cylinders")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("MPG by Number of Cylinders")
plt.suptitle("")
plt.show()
```

```python
#Qn 3

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
df_hp = df.dropna(subset=["horsepower"])
plt.scatter(df_hp["horsepower"], df_hp["mpg"], alpha=0.7)
plt.xlabel("Horsepower")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Horsepower vs. MPG")
plt.show()
```

```python
# Qn 4

import numpy as np

plt.scatter(df["weight"], df["mpg"], alpha=0.7, label="Data points")
z = np.polyfit(df["weight"], df["mpg"], 1)
p = np.poly1d(z)
plt.plot(df["weight"], p(df["weight"]), color="red", linewidth=2, label="Trend line")
plt.xlabel("Weight")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Weight vs. MPG with Trend Line")
plt.legend()
plt.show()
```

```python
# Qn 5

avg_mpg_per_year = df.groupby("model year")["mpg"].mean()
plt.plot(avg_mpg_per_year.index, avg_mpg_per_year.values, marker="o")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.title("Average MPG Across Model Years")
plt.grid(True)
plt.show()
```

```python
# Qn 6

origin_counts = df["origin"].value_counts().sort_index()
origin_labels = {1: "USA", 2: "Europe", 3: "Japan"}
plt.bar(origin_counts.index.map(origin_labels), origin_counts.values)
plt.xlabel("Origin")
plt.ylabel("Number of Cars")
plt.title("Number of Cars by Origin")
plt.show()
```

```python
# Qn 7

df.boxplot(column="acceleration", by="cylinders", grid=False)
plt.xlabel("Number of Cylinders")
plt.ylabel("Acceleration")
plt.title("Acceleration by Number of Cylinders")
plt.suptitle("")
plt.show()
```

```python
# Qn 8

car_counts_by_year = df["model year"].value_counts().sort_index()
plt.bar(car_counts_by_year.index, car_counts_by_year.values)
plt.xlabel("Model Year")
plt.ylabel("Number of Cars")
plt.title("Number of Cars by Model Year")
plt.show()

most_cars_year = car_counts_by_year.idxmax()
most_cars_count = car_counts_by_year.max()
print(
    f"The year with the most car entries is {most_cars_year} with {most_cars_count} cars."
)
```

```python
# Qn 9

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(
    df_hp["weight"],
    df_hp["horsepower"],
    df_hp["mpg"],
    c=df_hp["mpg"],
    cmap="viridis",
    alpha=0.7,
)
ax.set_xlabel("Weight")
ax.set_ylabel("Horsepower")
ax.set_zlabel("MPG")
ax.set_title("3D Scatter Plot: Weight vs Horsepower vs MPG")
plt.show()
```

```python
# Qn 10
top10_cars = (
    df.nlargest(10, "mpg")[["car name", "mpg"]].set_index("car name").sort_values("mpg")
)
top10_cars["mpg"].plot(kind="barh", color="skyblue")
plt.xlabel("Miles per Gallon (mpg)")
plt.title("Top 10 Cars with Highest MPG")
plt.tight_layout()
plt.show()
```
